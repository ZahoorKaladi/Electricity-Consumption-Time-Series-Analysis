```python
df.dropna(inplace=True)
df
```

```python
df=df.rename(columns={'Date_Time':'Date Time','Usage_kW':'Total Usage KW','AC_DR_kW':'AC_DRoom','UPS_kW':'UPS','LR_kW':'LivingRoom','Kitchen_kW':'Kitchen','AC_BR_kW':'AC Bedroom'})
df
```

```python
df.drop("AC_Dr_kW",axis=1,inplace=True)
```

```python
dfDate Time	Total Usage KW	AC_DRoom	UPS	LivingRoom	Kitchen	AC_Dr_kW	AC Bedroom
```

```python
# Convert 'Date Time' to datetime format
df['Date Time'] = pd.to_datetime(df['Date Time'], format='%Y-%m-%d %H:%M:%S')  # Adjust format if needed

# Drop rows with any missing values
df.dropna(inplace=True)

# Set 'Date Time' as the index
df.set_index('Date Time', inplace=True)

# Verify the changes
df.head()
```

```python
# Plot the total electricity usage over time
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Total Usage KW'], label='Total Usage (KW)')
plt.title('Electricity Consumption over Time')
plt.xlabel('Date Time')
plt.ylabel('Total Usage (KW)')
plt.legend()
plt.grid(True)
plt.show()

# You can also plot the usage for different rooms
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['AC_DRoom'], label='AC in Dining Room')
plt.plot(df.index, df['UPS'], label='UPS Usage')
plt.plot(df.index, df['LivingRoom'], label='Living Room')
plt.plot(df.index, df['Kitchen'], label='Kitchen')
plt.plot(df.index, df['AC_Dr_kW'], label='AC in D Room KW')
plt.plot(df.index, df['AC Bedroom'], label='AC in Bedroom')
plt.title('Electricity Consumption by Different Areas')
plt.xlabel('Date Time')
plt.ylabel('Usage (KW)')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Resample the data to get daily total consumption
df_daily = df.resample('D').sum()  # Sum the consumption for each day

# Plot the daily consumption
plt.figure(figsize=(10, 6))
plt.plot(df_daily.index, df_daily['Total Usage KW'], label='Total Usage (Daily)', color='orange')
plt.title('Daily Total Electricity Consumption')
plt.xlabel('Date')
plt.ylabel('Total Usage (KW)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Display summary statistics
df.describe()

# Check for any trends or seasonal patterns
# You can plot rolling averages to observe trends
df['Total Usage KW'].rolling(window=30).mean().plot(figsize=(10, 6), label='30-Day Rolling Average', color='red')
plt.plot(df.index, df['Total Usage KW'], label='Original Data', alpha=0.5)
plt.title('Total Electricity Usage with 30-Day Rolling Average')
plt.xlabel('Date Time')
plt.ylabel('Total Usage (KW)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA model
model = ARIMA(df_daily['Total Usage KW'], order=(5, 1, 0))  # Adjust parameters as needed
model_fit = model.fit()

# Make predictions
forecast = model_fit.forecast(steps=7)  # Forecast next 7 days
forecast_index = pd.date_range(start=df_daily.index[-1], periods=8, freq='D')[1:]

# Plot the forecast
plt.figure(figsize=(10, 6))
plt.plot(df_daily.index, df_daily['Total Usage KW'], label='Historical Data')
plt.plot(forecast_index, forecast, label='Forecast', color='red')
plt.title('Electricity Consumption Forecast (Next 7 Days)')
plt.xlabel('Date')
plt.ylabel('Total Usage (KW)')
plt.legend()
plt.grid(True)
plt.show()
```
